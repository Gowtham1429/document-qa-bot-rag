```python
from pypdf import PdfReader

pdf_file = "IELTS-Writing-Tips-Workbook.pdf"

reader = PdfReader(pdf_file)

all_text = ""

for page in reader.pages:
    text = page.extract_text()
    if text:
        all_text += text + "\n"

print("Total characters:", len(all_text))
print("\nFirst 1000 characters:\n")
print(all_text[:1000])
```

```python
def chunk_text(text, chunk_size=1000, overlap=200):
    chunks = []

    start = 0

    while start < len(text):
        end = start + chunk_size

        chunks.append(text[start:end])

        start += (chunk_size - overlap)

    return chunks


chunks = chunk_text(all_text)

print("Total Chunks:", len(chunks))
print("\nFirst Chunk:\n")
print(chunks[0][:500])
```

```python
import chromadb

client = chromadb.Client()

collection = client.get_or_create_collection(
    name="ielts_documents"
)

for i, chunk in enumerate(chunks):

    embedding = genai.embed_content(
        model="models/gemini-embedding-001",
        content=chunk
    )["embedding"]

    collection.add(
        ids=[f"chunk_{i}"],
        documents=[chunk],
        embeddings=[embedding]
    )

    if i % 20 == 0:
        print(f"Indexed {i} chunks...")

print("Finished indexing!")
```

```python
query_embedding = genai.embed_content(
    model="models/gemini-embedding-001",
    content="How can I improve IELTS Writing?"
)["embedding"]

results = collection.query(
    query_embeddings=[query_embedding],
    n_results=3
)

for i, doc in enumerate(results["documents"][0], start=1):
    print(f"\nResult {i}")
    print("-"*50)
    print(doc[:500])
```

```python
import google.generativeai as genai

def ask_document(question):

    query_embedding = genai.embed_content(
        model="models/gemini-embedding-001",
        content=question
    )["embedding"]

    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=3
    )

    context = "\n\n".join(results["documents"][0])

    prompt = f"""
You are a helpful document assistant.

Answer ONLY from the provided context.

CONTEXT:
{context}

QUESTION:
{question}
"""

    model = genai.GenerativeModel("models/gemini-3.5-flash")

    response = model.generate_content(prompt)

    return response.text
```

```python
answer = ask_document(
    "What are the common mistakes that reduce IELTS Writing scores?"
)

print(answer)
```
